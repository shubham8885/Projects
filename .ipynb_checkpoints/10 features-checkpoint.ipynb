```python

# Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

from sklearn.ensemble import AdaBoostClassifier, BaggingClassifier
from xgboost import XGBClassifier
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
StandardScaler = StandardScaler()  
columns_to_scale = ['age','sex','cp','trestbps','chol','fbs','restecg','thalach','exang','oldpeak','slope','ca','thal'] 
df[columns_to_scale] = StandardScaler.fit_transform(df[columns_to_scale])
```

```python
# Apply PCA
pca = PCA()
pca.fit(X)

# Get the names of the features
feature_names = X.columns

# Get the PCA scores for each feature
pca_scores = pca.explained_variance_ratio_

# Create DataFrame with feature names and their PCA scores
pca_results = pd.DataFrame({
    'Features': feature_names,
    'Score': pca_scores
})

# Display the DataFrame
print("Score of all features by PCA FE.")
print(pca_results)
```

```python
# Sort the DataFrame by 'Score' column in descending order
pca_results_sorted = pca_results.sort_values(by='Score', ascending=False)

# Select the top seven features
selected_features = pca_results_sorted.head(10)

# Display the selected features
print("Selected Features:")
print(selected_features)
```

```python
# List of all features
all_features = pca_results['Features'].tolist()

# List of selected features
selected_feature_names = selected_features['Features'].tolist()

# Remaining features
remaining_features = [feature for feature in all_features if feature not in selected_feature_names]

# Display the remaining features
print("Remaining Features:")
for feature in remaining_features:
    print(feature)
```

```python
X= df
y= df['target']
```

```python
#X= df.drop(['target','thalach','exang','oldpeak','slope','ca','thal'], axis=1) 
#y= df['target']
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
# Scale the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
#check those dataset
print('X_train-', X_train.size)
print('X_test-',X_test.size)
print('y_train-', y_train.size)
print('y_test-', y_test.size)
```

```python

# Apply PCA
pca = PCA(n_components=10)  
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)
```

```python
# Define classifiers
classifiers = {
    "Logistic Regression": LogisticRegression(random_state=42),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42),
    "SVM": SVC(random_state=42),
    "Naive Bayes": GaussianNB(),
    "KNN": KNeighborsClassifier(),
    "AdaBoost": AdaBoostClassifier(random_state=42),
    "Bagging": BaggingClassifier(random_state=42),
    "XGBoost": XGBClassifier(random_state=42)

}
```

```python
accuracy_scores = []  
sensitivity_scores = []
specificity_scores = []
```

```python
from sklearn.metrics import confusion_matrix

for name, clf in classifiers.items():
    clf.fit(X_train_pca, y_train)
    y_pred = clf.predict(X_test_pca)
    accuracy = accuracy_score(y_test, y_pred) * 100
    tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
    sensitivity = tp / (tp + fn)  # Sensitivity = True Positive Rate
    specificity = tn / (tn + fp)  # Specificity = True Negative Rate
    sensitivity_scores.append(sensitivity)
    specificity_scores.append(specificity)
    print(f"{name} Accuracy: {accuracy:.2f}% | Sensitivity: {sensitivity:.2f} | Specificity: {specificity:.2f}")
```

```python
import pandas as pd

# Initialize lists to store scores
accuracy_scores = []
sensitivity_scores = []
specificity_scores = []

# Loop through classifiers and calculate scores
for name, clf in classifiers.items():
    clf.fit(X_train_pca, y_train)
    y_pred = clf.predict(X_test_pca)
    accuracy = accuracy_score(y_test, y_pred) * 100
    tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
    sensitivity = tp / (tp + fn) *100  # Sensitivity = True Positive Rate
    specificity = tn / (tn + fp)  *100 # Specificity = True Negative Rate
    
    # Append scores to respective lists
    accuracy_scores.append(accuracy)
    sensitivity_scores.append(sensitivity)
    specificity_scores.append(specificity)

# Create DataFrame to store scores
scores_df = pd.DataFrame({
    'Classifier': list(classifiers.keys()),
    'Accuracy (%)': accuracy_scores,
    'Sensitivity': sensitivity_scores,
    'Specificity': specificity_scores
})

# Display the DataFrame
print(scores_df)
```
